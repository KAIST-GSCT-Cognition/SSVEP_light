```python
import os
import numpy as np
import json
```

```python
trs, vas = [], []

hzs = ["7hz","10hz","12hz", "all"]
platforms = ["Sc", "VR"]
types = ["fold_wise", "dependent"]

header = ['Model','Feature','Data','Hz','User_independent_Train','User_independent_Eval','User_dependent_Train','User_dependent_Eval']
results = []
for hz in hzs:
    for plt in platforms:
        with open(f"../ssvl/act_cls/svm/exp/{plt}_psd_{hz}_results.json", "r") as f:
            json_data = json.load(f)

            res = ['SVM','pcb',plt,hz]
            for t in types:
                res.append(json_data[t]['train_score'])
                res.append(json_data[t]['test_score'])
            results.append(res)
```

```python
import pandas as pd

df = pd.DataFrame(results, columns=header)
df
```
